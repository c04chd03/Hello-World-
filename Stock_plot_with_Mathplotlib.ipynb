We start by installing and loading the yahoo finance module in commnad line.

```python
!pip install yfinance
import yfinance
```

Then we load the data for specific stock. following the sequence of the yfinance.download module.
In this case we took the ticker, the starting date to gather tha data from and the ending date, including an auto adjustment to adjust the data automatically.

```python
yfinance.download("NVDA",
                  "2010-01-01",
                  "2024-01-01",
                  auto_adjust=True)
```

Then we narrow down the data by selecting only the close value for the given stock, and we use [] twice  because the dataframe is a 2*2 matrix, and we need only the close column (otherwise we will limit the date as well in the outer [] pair.

```python
dataframe = yfinance.download("NVDA",
                  "2010-01-01",
                  "2024-01-01",
                  auto_adjust=True)

dataframe = dataframe [["Close"]]
dataframe
```

we add dropna to drop the empty spaces or nod valid values.
In the result we can see the number of rows is the same as before. no NA in the set.

```python
dataframe = dataframe.dropna()
dataframe
```

import mathplotlib and pyplot module and named as pyplot for short.

```python
import matplotlib.pyplot as pyplot
```

We use pyplot to plot the close values for the stock by defining the color and the size of the plot, figsize uses x,y references. Color uses the initial letter for each color or the collor name.

```python
dataframe.Close.plot(figsize=(25,7),
                     color="gray")
pyplot.title("Nvidia Stock 2010-2024")
pyplot.ylabel("Nvidia Stock Value")
```
